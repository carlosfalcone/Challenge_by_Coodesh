# Carregamento dos dataframes

```python
# Carregamento e criaÃ§Ã£o dos dataframes

df_netflix = pd.read_csv('C:/Users/Falcone/Documents/PastaDrive/ProcessosCadidatura/Coodesh/AnaliseTitulosNetflixAmazon/data/netflix_titles.csv')
df_amazon = pd.read_csv('C:/Users/Falcone/Documents/PastaDrive/ProcessosCadidatura/Coodesh/AnaliseTitulosNetflixAmazon/data/amazon_prime_titles.csv')
df_all = pd.concat([df_netflix, df_amazon])
print(f'Quantidade de linhas -  df_neflix: {len(df_netflix)}, df_amazon: {len(df_amazon)}, df_all: {len(df_all)}')
print('\n\n                                         ____Tabela com TODOS os dados____')
df_all
```

# QuestÃ£o 1 - Top 10 atores/atrizes considerando todos os dados

```python
## TODOS OS DADOS - ALL

# Limpeza dos dados - Preencher valores NaN com string vazia
df_all1['cast'] = df_all1['cast'].fillna('')  

# SeparaÃ§Ã£o dos artistas informados na coluna 'cast' utilizando o separador virgula (,) 
# CriaÃ§ao de listas de artistas para cada linha
lista_artistas = df_all1['cast'].str.split(', ')

# CriaÃ§Ã£o de uma lista Ãºnica com todos os artistas 
todos_artistas = []
for sublista in lista_artistas:
    for artista in sublista:
        if artista.strip() != '':
            todos_artistas.append(artista)
            
# ExpressÃ£o regular para filtrar apenas os nomes (sem nÃºmeros), pois foram encontrados alguns nÃºmeros 
# nas informaÃ§oes da coluna 'cast'
somente_artistas = []
for nome in todos_artistas:
    if re.sub(r'\d', '', nome).strip() != '':
        somente_artistas.append(re.sub(r'\d', '', nome).strip())

# TransformaÃ§ao do tipo lista em series
series_artistas = pd.Series(somente_artistas)

# 10 artistas mais famosos (maior numero de vezes encontrado na lista participaÃ§ao)
contagem_series_artistas = series_artistas.value_counts()
contagem_series_artistas.head(10)
```

```python
## TODOS OS DADOS - ALL
# Limpeza dos dados - Preencher valores NaN com string vazia
df_all1['country'] = df_all1['country'].fillna('')  

# SeparaÃ§Ã£o dos paÃ­ses informados na coluna 'country' utilizando o separador virgula (,) 
# CriaÃ§ao de listas de paÃ­ses para cada linha
lista_paises = df_all1['country'].str.split(', ')

# CriaÃ§Ã£o de uma lista Ãºnica com todos os paÃ­ses 
todos_paises = []
for sublista in lista_paises:
    for pais in sublista:
        if pais.strip() != '':
            todos_paises.append(pais)
            
# TransformaÃ§ao do tipo lista em series
series_paises = pd.Series(todos_paises)

# 10 artistas mais famosos (maior numero de vezes encontrado na lista participaÃ§ao)
contagem_series_paises = series_paises.value_counts()
print('Top 5 paises produtores considerando TODOS os dados:')
print(contagem_series_paises.head(5))


## NETFLIX
# Limpeza dos dados - Preencher valores NaN com string vazia
df_netflix1['country'] = df_netflix1['country'].fillna('')  

# SeparaÃ§Ã£o dos paÃ­ses informados na coluna 'country' utilizando o separador virgula (,) 
# CriaÃ§ao de listas de paÃ­ses para cada linha
lista_paises_netflix = df_netflix1['country'].str.split(', ')

# CriaÃ§Ã£o de uma lista Ãºnica com todos os paÃ­ses 
todos_paises_netflix = []
for sublista in lista_paises_netflix:
    for pais in sublista:
        if pais.strip() != '':
            todos_paises_netflix.append(pais)
            
# TransformaÃ§ao do tipo lista em series
series_paises_netflix = pd.Series(todos_paises_netflix)

# 10 artistas mais famosos (maior numero de vezes encontrado na lista participaÃ§ao)
contagem_series_paises_netflix = series_paises_netflix.value_counts()
print('\nTop 5 paises produtores considerando somente os dados da NETFLIX:')
print(contagem_series_paises_netflix.head(5))


## AMAZON
# Limpeza dos dados - Preencher valores NaN com string vazia
df_amazon1['country'] = df_amazon1['country'].fillna('')  

# SeparaÃ§Ã£o dos paÃ­ses informados na coluna 'country' utilizando o separador virgula (,) 
# CriaÃ§ao de listas de paÃ­ses para cada linha
lista_paises_amazon = df_amazon1['country'].str.split(', ')

# CriaÃ§Ã£o de uma lista Ãºnica com todos os paÃ­ses 
todos_paises_amazon = []
for sublista in lista_paises_amazon:
    for pais in sublista:
        if pais.strip() != '':
            todos_paises_amazon.append(pais)
            
# TransformaÃ§ao do tipo lista em series
series_paises_amazon = pd.Series(todos_paises_amazon)

# 10 artistas mais famosos (maior numero de vezes encontrado na lista participaÃ§ao)
contagem_series_paises_amazon = series_paises_amazon.value_counts()
print('\nTop 5 paises produtores considerando somente os dados da AMAZON:')
print(contagem_series_paises_amazon.head(5))
```

# Questao 3 - MÃªs no qual hÃ¡ mais adiÃ§Ãµes de filmes na plataforma Netflix

```python
## NETFLIX

# Limpeza dos dados - Preencher valores NaN com string vazia
df_netflix1['date_added'] = df_netflix1['date_added'].fillna('')  

# Filtrar apenas por filmes
df_movies_netflix = df_netflix1.loc[df_netflix1['type'] == 'Movie']

# SeparaÃ§Ã£o dos meses informados na coluna 'date_added' utilizando o separador virgula (,) 
# CriaÃ§ao de listas de datas para cada linha
lista_datas_netflix = df_movies_netflix['date_added'].str.split(', ')

# CriaÃ§Ã£o de uma lista Ãºnica com todos as datas
todas_datas_netflix = []
for sublista in lista_datas_netflix:
    for data in sublista:
        if data.strip() != '':
            todas_datas_netflix.append(data)

# ExpressÃ£o regular para filtrar apenas os meses (sem nÃºmeros)
somente_meses_netflix = []
for nome in todas_datas_netflix:
    if re.sub(r'\d', '', nome).strip() != '':
        somente_meses_netflix.append(re.sub(r'\d', '', nome).strip())
            
# TransformaÃ§ao do tipo lista em series
series_meses_netflix = pd.Series(somente_meses_netflix)

# MÃªs com maior produÃ§ao de FILMES
contagem_series_meses_netflix = series_meses_netflix.value_counts()
print('MES no qual hÃ¡ mais adiÃ§Ãµes de FILMES na plataforma Netflix:')
print(contagem_series_meses_netflix.head(5).index[0])
```
